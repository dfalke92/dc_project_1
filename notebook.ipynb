```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
print(netflix_df)
```

```python
# Filter the data for movies released in the 1990s

movies_only = netflix_df['type'] == 'Movie'
netflix_df_movies = netflix_df[movies_only]

release_1990s = (netflix_df_movies['release_year'] >= 1990) & (netflix_df_movies['release_year'] < 2000)
netflix_df_movies_1990s = netflix_df_movies[release_1990s]
print('Number of Movies released between 1990 and 1999: ' + str(len(netflix_df_movies_1990s)))
```

```python
# Find the most frequent movie duration

# mean and median for reference
print('Mean: ' + str(np.mean(netflix_df_movies_1990s['duration'])) + ' mins')
print('Median: ' + str(np.median(netflix_df_movies_1990s['duration'])) + ' mins')

# histogram of all cases
plt.hist(netflix_df_movies_1990s['duration'], bins=len(netflix_df_movies_1990s))
plt.title('Duration span of movies in the 1990s')
plt.xlabel('duration (in min)')
plt.ylabel('counts (in total)')
plt.show()

plt.clf()

# histogram of only the most frequent cases for more approximate answer
max_frequent_durations = (netflix_df_movies_1990s['duration'] >= 85) & (netflix_df_movies_1990s['duration'] <= 115)
netflix_df_movies_1990s_max = netflix_df_movies_1990s[max_frequent_durations]
plt.hist(netflix_df_movies_1990s_max['duration'], bins=len(netflix_df_movies_1990s_max))
plt.title('Duration span of movies in the 1990s (most frequent cases)')
plt.xlabel('duration (in min)')
plt.ylabel('counts (in total)')
plt.show()

# first histogram shows highest count around 90
# second histogram proves highest count at 94
print('Most frequent movie duration taken from histrogram: 94')
duration = 94
```
